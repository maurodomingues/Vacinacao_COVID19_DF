```python
import os
import numpy as np
import pandas as pd
import openpyxl
```

## Conjunto de Dados - CORONAVÍRUS BRASIL
https://covid.saude.gov.br/

```python
diretorio  = 'Dataset/hist'
```

```python
df_dados          = pd.DataFrame()    
df_df = pd.DataFrame()
lista_arquivos    = []

# Lista os arquivos do diretório
for arquivo in os.listdir(diretorio):
    if arquivo.endswith('.csv'):
       lista_arquivos.append(arquivo) 
       
# Carrega o DataFrame 
for dataset in lista_arquivos:
    caminho_arquivo   = os.path.join(diretorio, dataset)
    df_dados          = pd.read_csv(caminho_arquivo, sep=';', low_memory=False)
    df_df = pd.concat([df_df, df_dados], ignore_index=True)
```

```python
df_df.info()
```

```python
df_fora_do_DF = df_df[df_df['estado'] != 'DF'].index
df_df       = df_df.drop(df_fora_do_DF)
```

```python
df_df.isnull().sum()
```

```python
df_df.dropna(subset=['nomeRegiaoSaude'], inplace = True)
```

```python
df_df
```

```python
df_int = df_df[['data','casosAcumulado','obitosAcumulado']]
```

```python
df_int.loc[:, 'data'] = pd.to_datetime(df_int['data'])
```

```python
df_int.loc[:,'casosAcumulado'] = df_int['casosAcumulado'].astype(int)
```
